# Collaborative Filtering Based Recommender System

1. Memory-Based Collaborative Filtering approaches can be divided into two main sections: 

- user-user filtering takes a particular user, find users that are similar to that user based on similarity of ratings, and recommend items that those similar users liked. 

“Users who are similar to you also liked …”

- item-item filtering will take an item, find users who liked that item, and find other items that those users or similar users also liked. It takes items and outputs other items as recommendations. makes recommendations based on user-product interaction in the past. the assuumtion behind the algorithm is that users like similar products and dislike similar products so they give similar ratings to similar products
“Users who liked this item also liked …”

in practice, item-item outperforms user-user in many cases

Item-item algo: 
- for item i, find other similar items
- esmitate rating for item i based on rating for similar items
- from item i, compute the similarity between item i and every items (restrict to items have been rated by user X)
- take the top k of those as neighborhoods 

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200 # users rate more than 200 books 
padhe_likhe_users = x[x].index # user indexes who have rated more than 200 movies
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50 # books have more than 50 rates
famous_books = y[y].index 
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)] # books have been rated by more than 50 users
```

create a user-product matrix

```python
user_item = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')

user_item.fillna(0,inplace=True)
```

```python
user_item
```

```python
# Create a pivot table with User-ID as the index and Book-Title as the columns
pt = final_ratings.pivot_table(index='User-ID', columns='Book-Title', values='Book-Rating')

# Fill the missing values with 0
pt.fillna(0, inplace=True)

# Calculate the similarity among items
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt.T) # Transpose the pivot table to get book names as columns and indexes

# Convert the similarity matrix into a pandas DataFrame with book names as columns and indexes
similarity_df = pd.DataFrame(similarity_scores, index=pt.columns, columns=pt.columns)
```

```python
similarity_df
```

Prediction:

Create a list of the movies that user i has read and rated.

Rank the similarities between the movies user i rated and movie j.

Select top n movies with highest similarity scores.

Calculate the predicted rating using weighted average of similarity scores and the ratings from user 1.

```python
# Item-based recommendation function
import operator
def item_based_rec(picked_userid=2276, number_of_similar_items=5, number_of_recommendations =3):
    #Create a list of the books that user i has read and unread.
    picked_userid_read_unread = pd.DataFrame(user_item[[picked_userid]])

    # convert the index to a new column 'Book-Title'
    picked_userid_read_unread = picked_userid_read_unread.rename_axis('Book-Title').reset_index()
    # rename column named 2276 to 'Rating'
    picked_userid_read_unread = picked_userid_read_unread.rename(columns={picked_userid: 'Rating'}).sort_values(ascending=False,by='Rating')

    #Create a list of the books that user i has read and rated.
    picked_userid_read = picked_userid_read_unread.loc[picked_userid_read_unread['Rating']!=0]

    # Books that the target user has not read
    picked_userid_unread = picked_userid_read_unread.loc[picked_userid_read_unread['Rating']==0]

    # Dictionary to save the unread books and predicted rating pair
    rating_prediction ={}  
    # Loop through unread books          
    for picked_book in picked_userid_unread['Book-Title']: 

        # Calculate the similarity score of the picked book with other books
        picked_book_similarity_score = similarity_df[[picked_book]].reset_index().rename(columns={picked_book:'similarity_score'})
    
        # Rank the similarities between the picked user read books and the picked unread book.

        picked_userid_read_similarity = pd.merge(left=picked_userid_read, 
                                                    right=picked_book_similarity_score, 
                                                    on='Book-Title', 
                                                    how='inner')\
                                        .sort_values('similarity_score', ascending=False)[:number_of_similar_items]
        # Calculate the predicted rating using weighted average of similarity scores and the ratings from user 1
        predicted_rating = round(np.average(picked_userid_read_similarity['Rating'], 
                                            weights=picked_userid_read_similarity['similarity_score']), 6)
        # Save the predicted rating in the dictionary
        rating_prediction[picked_book] = predicted_rating
    # Return the top recommended movies
    return sorted(rating_prediction.items(), key=operator.itemgetter(1), reverse=True)[:number_of_recommendations]
# Get recommendations
recommended_books = item_based_rec(picked_userid=271705, number_of_similar_items=5, number_of_recommendations =3)
recommended_books
```

2. model based approach
use machine learning to predict unpredicted items

In the user-item matrix, there are two dimensions: The number of users The number of items

If the matrix is mostly empty, reducing dimensions can improve the performance of the algorithm in terms of both space and time. Can use various methods like matrix factorization or autoencoders to do this.

Matrix Factorization (MF): The idea behind such models is that attitudes or preferences of a user can be determined by a small number of hidden factors. We can call these factors as Embeddings.

Matrix decomposition can be reformulated as an optimization problem with loss functions and constraints. 

```python
users_books_pivot_matrix_df = pt.values
```

```python
users_books_pivot_matrix_df
```

```python
from scipy.sparse.linalg import svds

NUMBER_OF_FACTORS_MF = 15

#Performs matrix factorization of the original user item matrix
U, sigma, Vt = svds(users_books_pivot_matrix_df, k = NUMBER_OF_FACTORS_MF)
```

```python
sigma = np.diag(sigma)
sigma.shape
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 
all_user_predicted_ratings
```
